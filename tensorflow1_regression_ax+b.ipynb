# tensorflow를 사용한 회귀식

```python
import tensorflow as tf
tf.__version__
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
learning_rate = 0.001
display_step = 100
training_epoch = 30000
```

```python
x = tf.placeholder(tf.float32, shape=[1, None])
y = tf.placeholder(tf.float32, shape=[1, None])
```

```python
a = tf.Variable(tf.ones([1,1]))
b = tf.Variable(tf.ones([1,1]))
```

```python
y_estimated = tf.add(tf.matmul(a, x), b)
```

```python
error = tf.reduce_mean(tf.square(y - y_estimated))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate)
```

```python
model = optimizer.minimize(error)
```

```python
sess = tf.Session()
```

```python
init = tf.global_variables_initializer()
```

```python
sess.run(init)
```

```python
for i in range(0,  training_epoch):
    sess.run(model, feed_dict={x: x_input, y: y_true})
    if ((i % display_step) == 0):
        error_out = sess.run(error, feed_dict={x: x_input, y: y_true})
        a_out = sess.run(a, feed_dict={x: x_input, y: y_true})
        b_out = sess.run(b, feed_dict={x: x_input, y: y_true})
        print(i, "error = ", error_out, "  a = ", a_out, "  b = ", b_out)
```
